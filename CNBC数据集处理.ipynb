```python
import os
from PIL import Image
dir='/Users/liangxiaoyun/Downloads/program/Face_Style_Transfer_Program/folders/original_data/CNBC'
files = os.listdir(dir)
#files.remove(files[12])
files.remove('.DS_Store')
files
```

```python
save_path="/Users/liangxiaoyun/Downloads/program/Face_Style_Transfer_Program/stargan-master/data/KDEF/train/"
for name in files:
    folder=os.path.join(dir,name)
    if os.path.isdir(folder):  #判断是文件夹还是文件
        imagenames = os.listdir(folder)
        for imagename in imagenames:
            imagepth=os.path.join(folder,imagename)
            img = Image.open(imagepth)
            imgnameli=imagename.split('_')
            #print(len(imgnameli[2]))#6:emotion;7:angle.   only use emotion
            if len(imgnameli[2])==6:     
                emotion=imgnameli[2][:2]
                print(emotion)
                if emotion =='FE':
                    img.save(save_path+'afraid/'+imagename)
                elif emotion=='AN':
                    img.save(save_path+'angry/'+imagename)
                elif emotion=='DI':
                    img.save(save_path+'disgusted/'+imagename)
                elif emotion=='HA':
                    img.save(save_path+'happy/'+imagename)
                elif emotion=='NE':
                    img.save(save_path+'neutral/'+imagename)
                elif emotion=='SA':
                    img.save(save_path+'sad/'+imagename)
                elif emotion=='SU':
                    img.save(save_path+'surprised/'+imagename)
                else : # default, could also just omit condition or 'if True'
                    print("something else!")
                print(imagename+" has saved!")
```
